## Initialise user's account

```python
# Please set the following parameters.
name = "John Smith" # user name
portfolio_id = "AAA" # portfolio name (each user could have more than one portfolio)
date = "2015-10-20" # date of account
currency = "EUR" # account currency
initial_balance = 130000
```
